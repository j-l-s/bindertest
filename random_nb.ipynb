```python
def nextRandom():
    global r
    r = (16807 * r + 1 ) % 32768
    return r
```

```python
def bit9(x):
    x = x >> 8
    return (x % 2)
```

```python
pair1 = 0
impair1 = 0
pair2 = 0
impair2 = 0
transition1 = -1;
transition2 = -1;
old1 = -1
old2 = -1
```

```python
for i in range(0, 1000000):
    rand = nextRandom()
#    print ("r = " + str(r))
    if rand %2 == 0:
        pair1 += 1
        pair = 1
    else:
        impair1 += 1
        pair = 0
    if old1 != pair:
        transition1 += 1
        old1 = pair

    if bit9(rand) == 0:
        pair2 += 1
        pile_ou_face = "pile"
        pair = 1
    else:
        impair2 += 1
        pile_ou_face = "face"
        pair = 0
#    print (pile_ou_face)
    if old2 != pair:
        transition2 += 1
        old2 = pair
```
